```python
#drop blank rows from dataset
presc_df=presc_df[presc_df['Presc01'].notna()]
presc_df
```

```python
#Create a list of lists
listoflists=[]
for i in range (0, 7501):
        listoflists.append([str(presc_df.values[i,j])
for j in range (0, 20)])
```

```python
#Feed list of lists into TransactionEncoder
TE=TransactionEncoder()
array=TE.fit(listoflists).transform(listoflists)
#Convert transaction encoder array back to dataframe
transaction=pd.DataFrame(array, columns= TE.columns_)
```

```python
#Drop the nan column in the dataframe
cleaned_presc= transaction.drop(['nan'], axis=1)
cleaned_presc.head(7501)
```
